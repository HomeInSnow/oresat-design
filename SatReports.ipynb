```python
import ephem
import numpy as np
import datetime
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
iss = "ISS (ZARYA)", "1 25544U 98067A   16145.69686492  .00003239  00000-0  55059-4 0  9992", "2 25544  51.6424 175.2139 0001726 113.6054 353.3802 15.54685316  1333"

start = datetime.datetime.now()
end = start + datetime.timedelta(7)
step = datetime.timedelta(0,1)

sat = ephem.readtle(*iss)
pdx = ephem.Observer()
pdx.lat = '45.5231'
pdx.lon = '122.6765'

critical_elevation = ephem.degrees(45)
```

```python
pass_happening = False
now = start

pass_list = []
elevations = []

while now < end:
    pdx.date = now
    sat.compute(pdx)
    alt = np.degrees(180/np.pi*sat.alt)
    
    if pass_happening:
        if alt < critical_elevation:
            pass_happening = False
            pass_end = now
            pass_list.append(((pass_end-pass_start).total_seconds(), pass_start, pass_end))
    else:
        if alt > critical_elevation:
            pass_happening = True
            pass_start = now
    elevations.append(sat.alt)
    now += step
```

```python
pass_lengths, pass_starts, pass_ends = zip(*pass_list)
pass_lengths = [x/60.0 for x in pass_lengths]


plt.figure(figsize=(15,10))
n, bins, patches = plt.hist(pass_lengths, 10)
plt.xlabel("Length of pass (minutes)")
plt.ylabel("Number of passes")
plt.title(u"ISS pass lengths over PDX for one week with critical elevation angle = %g°"%critical_elevation)
plt.show()
```
